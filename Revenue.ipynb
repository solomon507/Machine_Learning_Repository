```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
# Importing dataset
data = pd.read_csv(r"D:\Intellipaat\lms assignments\datasets\Revenue_prediction_model\revenue_prediction.csv")
data.head()
```

```python
#checking missing values
data.isna().sum()
```

```python
#dropping unnecessary columns
data = data.drop(columns=["Id","Name","Franchise","Category","City","No_Of_Item"])
data.head()
```

```python
#Independent feature and dependent variable
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#LinearRegression Model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#predicted data
y_pred = model.predict(x_test)
y_pred
```

```python
#visualization 
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,model.predict(x_train),color="green")
plt.title("Revenue vs orders")
plt.xlabel("orders")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_train,model.predict(x_train),color="green")
plt.title("Revenue vs orders")
plt.xlabel("orders")
plt.ylabel("Revenue")
plt.show()
```

```python
#r2_score
r2_score = r2_score(y_test,y_pred)
r2_score
```
